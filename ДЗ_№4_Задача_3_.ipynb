```python
import numpy as np

# Создаём вектора
vector1 = np.arange(0, 1.5, 0.1)
vector2 = np.arange(1, 7)

print(vector1)
print(vector2)
```

```python
# Создаём матрицы
matrix1 = vector1.reshape(5, 3)
matrix2 = vector2.reshape(3, 2)

print(matrix1)
print(matrix2)
```

```python
# Выполняем умножение матриц
result = np.dot(matrix1, matrix2)
print(result)
```
